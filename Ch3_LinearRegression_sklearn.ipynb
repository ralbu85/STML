## Import Dataset

```python
import pandas as pd

df = pd.read_csv('Boston.csv', index_col=0)
df
```

## Simple Linear Regression

```python
y = df['medv']
X = df[['lstat','age']]
```

```python
# import the scikit-learn linear-regression model
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression() # create the object
reg.fit(X,y) # learn the object from the data
```

```python
reg.predict(X) # predict
```

```python
reg.coef_
```

## Multiple Linear Regression

```python
y = df['medv']
X = df.drop(columns=['medv']) # Select entire vector for X
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

## Measure the goodness of fit

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# we need to make y_hat (prediction from only X)
y_pred = reg.predict(X)
```

```python
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
```

```python
mse, r2
```

## Linear Regression with transformation

```python
df = pd.read_csv('Carseats.csv')
df
```

```python
df2 = pd.get_dummies(df, columns=['ShelveLoc','Urban','US'], drop_first=True)
df2
```

```python
y = df2['Sales']
X = df2.drop(columns=['Sales'])
```

```python
X['Income_Advertising'] = X['Income'] * X['Advertising']
X['Price_Age'] = X['Price'] * X['Age']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
y_pred = reg.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y,y_pred)
r2 = r2_score(y,y_pred)
mse, r2
```
